```python
f = open("input8.txt", "r")
inp = f.read().strip().split('\n')
f.close()
```

```python
outputs = [t.strip().split(' ') for t in [s.split('|')[1] for s in inp]]
```

```python
def count1478(inputs):
    ones = 0
    fours = 0
    sevens = 0
    eights = 0
    for i in inputs:
        for j in i:
            if len(j) == 2:
                ones += 1
            elif len(j) == 4:
                fours += 1
            elif len(j) == 3:
                sevens += 1
            elif len(j) == 7:
                eights += 1
    return {'1':ones, '4':fours,'7':sevens,'8': eights}
```

```python
part_one = count1478(outputs)
```

```python
def decde_signal(code, key):
    # The easy ones:
    if len(code) == 2:
        key['1'] = code
    elif len(code) == 4:
        key['4'] = code
    elif len(code) == 3:
        key['7'] = code
    elif len(code) == 7:
        key['8'] = code
    # some logic required
    elif len(code) == 5:
        # 5
        if '6' in key:
            if all(c in key['6'] for c in code): # all(c in 'abcd' for c in 'abc') -- True
                key['5'] = code
        # 2
        if '1' in key:
            key_found = false
            for c in code:
                if c in key['1']:
                    key_found = true
            if not key_found:
                key['2'] = code
        # 3
        if '5' in key and '2' in key:
            if code != key['5'] and code != key['2']:
                key['3'] = code
    elif len(code) == 6:
        # 6
        # 9
        if '5' in key and '1' in key: # 5 & 1 in 9
            if all(c in code for c in key['5']) and all(c in code for c in key['1']):
                key['9'] = code
        # 0
             
    return key
```
