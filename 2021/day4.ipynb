### --- Day 4: Giant Squid ---
You're already almost 1.5km (almost a mile) below the surface of the ocean, already so deep that you can't see any sunlight. What you can see, however, is a giant squid that has attached itself to the outside of your submarine.

Maybe it wants to play bingo?

Bingo is played on a set of boards each consisting of a 5x5 grid of numbers. Numbers are chosen at random, and the chosen number is marked on all boards on which it appears. (Numbers may not appear on all boards.) If all numbers in any row or any column of a board are marked, that board wins. (Diagonals don't count.)

The submarine has a bingo subsystem to help passengers (currently, you and the giant squid) pass the time. It automatically generates a random order in which to draw numbers and a random set of boards (your puzzle input). For example:

```python
def part_one(numbers, board_arrays):
    for n in range(1,len(numbers)):
        called_nums = numbers[:n]
        for ba in board_arrays:
            ba2d = ba.reshape(5, 5)
            for m in range(0,5):
                column = ba2d[:,m]
                column.flatten().tolist()
                if all(nums in called_nums for nums in column):
                    board = ba
                    return part_one_score(board, called_nums)
                row = ba2d[m,:]
                row.flatten().tolist()
                if all(nums in called_nums for nums in row):
                    board = ba
                    return part_one_score(board, called_nums)
```

```python
def part_one_score(board, called_nums):
    unmarked = [num for num in board.flatten().tolist() if num not in called_nums]
    sum = 0
    for n in unmarked:
        sum = sum + n
    print(board,'\n\n','called:',called_nums,'\n','unmarked:',unmarked,'\n','sum:',sum,'last:', called_nums[-1], 'round:', len(called_nums))
    return sum * called_nums[-1]
```

```python
part_one(numbers, board_arrays)
```
